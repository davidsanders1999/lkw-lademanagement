```python
import pandas as pd
import matplotlib.pyplot as plt

df_intraday = pd.read_csv('/Users/davidsanders/Documents/GitHub/masterarbeit-flex-kpis/input/datenaufbereitung_intraday/intraday_2024_5min.csv', sep=';', decimal=',')
df_dayahead = pd.read_csv('/Users/davidsanders/Documents/GitHub/masterarbeit-flex-kpis/input/datenaufbereitung_dayahead/dayahead_2024_5min.csv', sep=';', decimal=',')
```

```python
df_intraday['Preis'] = df_intraday['Preis']/ 10
```

```python
datum = '2024-04-25'

df_intraday_filterd = df_intraday[df_intraday['Volles_Datum'].str.contains(datum)]
df_dayahead_filterd = df_dayahead[df_dayahead['Volles_Datum'].str.contains(datum)]

plt.plot(df_intraday_filterd['Minuten'], df_intraday_filterd['Preis'], label='Intraday')
plt.plot(df_dayahead_filterd['Minuten'], df_dayahead_filterd['Preis'], label='Dayahead')
```
