```python
import pandas as pd

data = pd.read_csv('/Users/davidsanders/Documents/GitHub/masterarbeit-flex-kpis/data/konf_optionen/konf_optionen_cl_2_quote_100-100-100_netz_100_pow_100-100-100_pause_45-540_M_1_Base.csv', sep=';', decimal=',')
data
```

```python
list_ladetype = data['Ladetype'].unique().tolist()

for lt in list_ladetype:
    max = data[data['Ladetype'] == lt]['Anzahl_Ladesaeulen'].max()
    data.loc[data['Ladetype'] == lt, 'Anzahl_Ladesaeulen_rel'] = data[data['Ladetype'] == lt]['Anzahl_Ladesaeulen'] / max
```

```python
data.to_csv('/Users/davidsanders/Documents/GitHub/masterarbeit-flex-kpis/data/konf_optionen/konf_optionen_cl_2_quote_100-100-100_netz_100_pow_100-100-100_pause_45-540_M_1_Base_rel_1.csv', sep=';', decimal='.', index=False)
```

```python
export_data = {'Ladetype': [], 'Anzahl_Ladesaeulen_rel': [], 'Ladequote': []}

for lt in list_ladetype:
    for i in range(1, 101):
        rel_anzahl_ladesaeulen = i / 100
        upper = data[(data['Ladetype'] == lt) & (data['Anzahl_Ladesaeulen_rel'] >= rel_anzahl_ladesaeulen)]['Anzahl_Ladesaeulen_rel'].min()
        # lower = data[(data['Ladetype'] == lt) & (data['Anzahl_Ladesaeulen_rel'] <= rel_anzahl_ladesaeulen)]['Anzahl_Ladesaeulen_rel'].max()
        rel_ladequote_upper = data[(data['Ladetype'] == lt) & (data['Anzahl_Ladesaeulen_rel'] == upper)]['Ladequote'].values[0]
        rel_ladequote =  rel_anzahl_ladesaeulen / upper * rel_ladequote_upper
        
        export_data['Ladetype'].append(lt)
        export_data['Anzahl_Ladesaeulen_rel'].append(rel_anzahl_ladesaeulen)
        export_data['Ladequote'].append(rel_ladequote)
                
export_data = pd.DataFrame(export_data)
export_data.head()

export_data.to_csv('/Users/davidsanders/Documents/GitHub/masterarbeit-flex-kpis/data/konf_optionen/konf_optionen_cl_2_quote_100-100-100_netz_100_pow_100-100-100_pause_45-540_M_1_Base_rel.csv', sep=';', index=False)
```
